```python
#Plotting a Pie Chart to show how many of the Fairies fit into each Type 1

fairies = pokemon_df[pokemon_df['Type 2'] == 'Fairy']

fairies['Type 1'].value_counts()
```

```python
fairies['Type 1'].unique()
```

```python
Fairy_Pokemons= ['Normal', 'Psychic', 'Water', 'Steel', 'Dragon', 'Grass', 'Electric', 'Rock']
 
data = [5, 6, 2, 3, 1, 2, 1, 3]

fig = plt.figure(figsize=(4, 4))
plt.pie(data, labels=Fairy_Pokemons)

plt.show()
```

```python
#Making it cuterrr

colors = ("pink", "violet", "mediumturquoise", "lightgray", "slateblue", "limegreen", "yellow", "goldenrod")

#wp is wedge property
wp = {'linewidth': 2, 'edgecolor': "white"}

fig, ax = plt.subplots(figsize=(4, 4))
wedges, texts = ax.pie(data,
                                  labels=Fairy_Pokemons,
                                  shadow=False,
                                  colors=colors,
                                  startangle=63,
                                  wedgeprops=wp,
                                  textprops=dict(color="black"))
```
